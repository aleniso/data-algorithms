```
import math
import matplotlib.pyplot as plt
import numpy as np


def solveQuadratic( a,b,c):
  #check for not solutions. Not quadratic equation
  if a  == 0:
    print("The equation: %+dX^2%+dX%+d is not quadratic" % (a,b,c))
    print("")
    return    

  #save the original value of variables to print later
  oData = [1,2,3]
  oData[0] = a
  oData[1] = b
  oData[2] = c

  #change the equation to form x^2+bx+c
  if a != 1 and a != 0:
      b = b/a
      c = c/a 
      a = a/a
  #calculate the average of the sum of the roots
  av = -b/2
  k = (av*av)-c

  #check for complex roots
  isComplex = None
  if k < 0: 
    k = k*(-1)
    isComplex = True

  #calculate the z value of (av - z)(av + z) = c
  z = math.sqrt(k)  

  if isComplex:
    print("Equation: %+dX^2%+dX%+d" % (oData[0],oData[1],oData[2]))
    if av-z == av+z:
      print("Root: (%f + %fi)" % (av,z) )
    else:
      print("Roots: (%f + %fi), (%f - %fi)" % (av,z,av,z) )

  else:
    print("Equation: %+dX^2%+dX%+d" % (oData[0],oData[1],oData[2]))
    if av-z == av+z:
      print("Root: (%f + %f)" % (av,z) )
      print("aproximate form: %f" % ((av+z)) )
    else:
      print("Roots: (%f + %f), (%f - %f)" % (av,z,av,z) )
      print("aproximate form: %f , %f" % ((av+z),(av-z)))

  print("")


solveQuadratic(1,-2,1)
solveQuadratic(1,-3,2)
solveQuadratic(1,1,-2)
solveQuadratic(1,-1,-2)

solveQuadratic(0,3,-10)
solveQuadratic(1,0,1)
solveQuadratic(0,0,-10)



    
```
