```
import cmath
import matplotlib.pyplot as plt
import numpy as np

#give the rational roots for fifth grade polynomial
#using synthetic division with the f != 0
#for f == 0 the zero root is return
def solveFifthPoly( a,b,c,d,e,f ):
  #check for correct equation
  if a == 0:
    print("Not fifth grade equation: %+dX^5%+dX^4%+dx^3%+d^2%+dx%+d " % (a,b,c,d,e,f))
    return
  #check for easy case
  if b == 0 and c == 0 and d == 0 and e == 0:
    if f < 0 and a > 0:
      root = math.pow( (-f/a), (1/5) )
      print("Equation: %+dX^5%+dX^4%+dx^3%+d^2%+dx%+d" % (a,b,c,d,e,f))
      print("Real root = %+f"%(root)) 
      return
    else:
      print("Equation: %+dX^5%+dX^4%+dx^3%+d^2%+dx%+d" % (a,b,c,d,e,f))
      print("No real solution")    
      return   

  #get the guesses for synthetic division
  #note: in general odd grade polynomial has at least one real root
  #however no necessarily rational
  if f != 0:
    #calculate f divisors
    divisors = []    
    assign = float(0)
    for ia in range(1, abs(f)+1):
      if (f % ia)==0:
        assign = float(ia)
        divisors.append(assign)
        divisors.append(-assign)

    #calculate a divisors
    assign = float(0)
    for ib in range(2, abs(a)+1):
      if (a % ib)==0:
        assign = float(ib)
        divisors.append(assign)
        divisors.append(-assign)

    if a <= f:
      #calculate a and f common divisors
      assign = float(0)
      for ic in range(2,abs(f)+1):
        if (a % ic) == 0 and (f % ic) == 0:  
          assign = float( ic )        
          divisors.append( assign )
          divisors.append( -assign )
    else:
      #calculate a and f common divisors
      assign = float(0)
      for ie in range(2,abs(a)+1):
        if (a % ie) == 0 and (f % ie) == 0:  
          assign = float(ie)        
          divisors.append( ie )
          divisors.append( -ie )

    #calculate the ratio of divisors of a,f and common 
    #divisors of a and f
    assign = float(0)
    for z in range(1, len(divisors)):
      assign = divisors[z]/z
      divisors.append(assign)
      divisors.append(-assign)

  else:
    print("Real Root: %d" % (f))
    return
  
  #erased duplicates
  divisors = list(set(divisors))

  #sythetic division for find rational roots
  #for fifth grade polynomial only
  roots = []  
  for i in range(0,len(divisors)-1):     
    k = divisors[i]   
    if (f+(k*e+((k**2)*d+((k**3)*c+((k**4)*b+(k**5)*a))))) == 0.0:      
      roots.append(k) 

  #print the result
  if len(roots) != 0:
    print("Equation: %+dX^5%+dX^4%+dx^3%+d^2%+dx%+d" % (a,b,c,d,e,f))
    for i in range(len(roots)):
      print("Rational roots: (%+f)" % (roots[i]) )
  else:
    print("Equation: %+dX^5%+dX^4%+dx^3%+d^2%+dx%+d" % (a,b,c,d,e,f))
    print("No rational roots by synthetic division found")


  print("")

solveFifthPoly(1,-15,85,-225,274,-120)
solveFifthPoly(1,-14,72,-166,167,-60)
```
